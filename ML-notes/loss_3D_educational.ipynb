```python
import numpy as np
import torch
from torch.functional import F
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
x = np.array([.5,.5])

w1 = np.linspace(-1,1,40)
w2 = np.linspace(-1,1,40)

grid = np.meshgrid(w1,w2)
```

```python
def f_loss(x,y):
    return np.pow( np.linalg.norm(x-y),2).mean(axis=0)
```

```python
y = F.sigmoid(torch.from_numpy( grid[0]*x[0] + grid[1]*x[1]))
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(grid[0], grid[1], y.numpy())

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

```python
fig = go.Figure(data=[go.Surface(x=grid[0], y=grid[1], z=y)])
fig.show()
```
